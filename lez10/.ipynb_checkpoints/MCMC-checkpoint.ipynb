```python
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
```

```python
# 0=cloud 1=sun
```

```python
def generate (today,p1,p2):
    tomorrow = np.random.uniform(0,1,1)
    if today == 0:
        if tomorrow<p1:
            return 0
        else:
            return 1
    if today == 1:
        if tomorrow<p2:
            return 0
        else:
            return 1
```

```python
p_cc = 0.5
p_sc = 0.5
p_ss = 0.9
p_cs = 0.1
```

```python
N = np.linspace(1,10**5,10**5,dtype=int) #number of days
cloud_days = np.zeros(len(N))
sun_days = np.zeros(len(N))

today = np.random.randint(0,2,1)  #I choose randomly the initial condition
for i in range(len(N)):
    if today == 0:
        cloud_days[i]=cloud_days[i-1]+1
        sun_days[i]=sun_days[i-1]
    if today == 1:
        sun_days[i]=sun_days[i-1]+1
        cloud_days[i]=cloud_days[i-1]
    today = generate(today,p_cc,p_cs)
```

```python
plt.plot(N,cloud_days/N)
plt.title('Cloudy days')
plt.xlabel('Days')
plt.ylabel('p(cloudy)')
plt.show()
plt.plot(N,sun_days/N)
plt.title('Sunny days')
plt.xlabel('Days')
plt.ylabel('p(sunny)')
plt.show()

#I discard the burn-in
burn_in = 1000
plt.plot(N[burn_in:],cloud_days[burn_in:]/N[burn_in:])
plt.title('Cloudy days. Burn-in = '+str(burn_in))
plt.xlabel('Days')
plt.ylabel('p(cloudy)')
plt.show()
plt.plot(N[burn_in:],sun_days[burn_in:]/N[burn_in:])
plt.title('Sunny days. Burn-in = '+str(burn_in))
plt.xlabel('Days')
plt.ylabel('p(sunny)')
plt.show()

burn_in = 10000
plt.plot(N[burn_in:],cloud_days[burn_in:]/N[burn_in:])
plt.title('Cloudy days. Burn-in = '+str(burn_in))
plt.xlabel('Days')
plt.ylabel('p(cloudy)')
plt.show()
plt.plot(N[burn_in:],sun_days[burn_in:]/N[burn_in:])
plt.title('Sunny days. Burn-in = '+str(burn_in))
plt.xlabel('Days')
plt.ylabel('p(sunny)')
plt.show()
```

```python
a1,b1,c1 = plt.hist(cloud_days[burn_in:]/N[burn_in:],100,density=True)
plt.title('Cloudy days')
plt.show()
a2,b2,c2 = plt.hist(sun_days[burn_in:]/N[burn_in:],100,density=True)
plt.title('Sunny days')
plt.show()
```

```python
print('Most probable p(clear) : '+str(b1[:-1][a1==np.max(a1)]))
print('Most probable p(sunny) : '+str(b2[:-1][a2==np.max(a2)]))
```
