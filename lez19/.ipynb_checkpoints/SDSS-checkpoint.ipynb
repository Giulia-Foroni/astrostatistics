```python
import sklearn
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import metrics
from astroML.utils import completeness_contamination
from sklearn.metrics import roc_curve

from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from astroML.classification import GMMBayes
from sklearn.neighbors import KNeighborsClassifier

import pandas as pd
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv("galaxyquasar.csv")
data
```

```python
fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['u']-data['g']).to_numpy(),y=(data['g']-data['r']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('u-g')
plt.ylabel('g-r')
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['u']-data['g']).to_numpy(),y=(data['r']-data['i']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('u-g')
plt.ylabel('r-i')
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['u']-data['g']).to_numpy(),y=(data['i']-data['z']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('u-g')
plt.ylabel('i-z')
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['g']-data['r']).to_numpy(),y=(data['r']-data['i']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('g-r')
plt.ylabel('r-i')
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['g']-data['r']).to_numpy(),y=(data['i']-data['z']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('g-r')
plt.ylabel('i-z')
plt.show()

fig = plt.figure(figsize=(4, 4))
sns.scatterplot(x=(data['r']-data['i']).to_numpy(),y=(data['i']-data['z']).to_numpy(),hue=data['class'].to_numpy())
plt.xlabel('r-i')
plt.ylabel('i-z')
plt.show()
```

```python
X_train, X_test = train_test_split(data,train_size=0.8)
```

```python
u_g_train = (X_train['u']-X_train['g']).to_numpy()
g_r_train = (X_train['g']-X_train['r']).to_numpy()
r_i_train = (X_train['r']-X_train['i']).to_numpy()
i_z_train = (X_train['i']-X_train['z']).to_numpy()
classes_train = (X_train['class']).to_numpy()
classes_train[classes_train == 'QSO'] = 1
classes_train[classes_train == 'GALAXY'] = 0
classes_train = np.float64(classes_train)

u_g_test = (X_test['u']-X_test['g']).to_numpy()
g_r_test = (X_test['g']-X_test['r']).to_numpy()
r_i_test = (X_test['r']-X_test['i']).to_numpy()
i_z_test = (X_test['i']-X_test['z']).to_numpy()
classes_test = (X_test['class']).to_numpy()
classes_test[classes_test == 'QSO'] = 1
classes_test[classes_test == 'GALAXY'] = 0
classes_test = np.float64(classes_test)
```

```python
matrix_train = np.column_stack((u_g_train,g_r_train,r_i_train,i_z_train))
matrix_test = np.column_stack((u_g_test,g_r_test,r_i_test,i_z_test))
```

# Gaussian Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(matrix_train,classes_train)
predicted_labels_gnb = gnb.predict(matrix_test)
```

```python
hue = np.array(predicted_labels_gnb,dtype=str)
hue[predicted_labels_gnb==0]='GALAXY'
hue[predicted_labels_gnb==1]='QSO'
sns.scatterplot(x=matrix_test[:,0],y=matrix_test[:,1],hue=hue)

'''
# predict the classification probabilities on a grid
xlim = (-2, 4)
ylim = (-7, 5)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = gnb.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)
plt.contour(xx, yy, Z, [0.5], colors='k')
'''

plt.xlabel('u-g')
plt.ylabel('g-r')
plt.show()
```

```python
print('Score = ', gnb.score(matrix_test,classes_test))
```

# K-Nearest neighbors

```python
neighbors = np.linspace(1,20,20,dtype = int)
model = KNeighborsClassifier()
grid = GridSearchCV(model, {'n_neighbors': neighbors}, scoring = 'accuracy', cv= 5) 
grid.fit(matrix_train,classes_train)
print('Best iperparameter: ', grid.best_params_)
print('Best score: ', grid.best_score_)
```

```python
knc = KNeighborsClassifier(9) # k=5
knc.fit(matrix_train,classes_train)
predicted_labels_knc = knc.predict(matrix_test)
```

# ROC curves

```python
scores = gnb.predict_proba(matrix_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(classes_test, scores)
plt.plot(fpr,tpr,label= 'GNB')

scores = lda.predict_proba(matrix_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(classes_test, scores)
plt.plot(fpr,tpr,label= 'LDA')

scores = qda.predict_proba(matrix_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(classes_test, scores)
plt.plot(fpr,tpr,label= 'QDA')

scores = gmmb.predict_proba(matrix_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(classes_test, scores)
plt.plot(fpr,tpr,label= 'GMMB')

scores = knc.predict_proba(matrix_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(classes_test, scores)
plt.plot(fpr,tpr,label= 'KNC')

plt.xlim(-0.001,0.04)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('ROC curves')
plt.legend()
plt.show()
```
