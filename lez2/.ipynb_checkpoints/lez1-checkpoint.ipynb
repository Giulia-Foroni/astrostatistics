```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
from numpy import zeros
import math
import statistics
```

```python
x_min = 0.1
x_max = 10
x = np.random.uniform(x_min,x_max,2000) 
p_x = 1/(x_max-x_min) #p(x)
```

```python
plt.hist(x,20,density=True)
plt.plot(np.linspace(np.min(x),np.max(x),100),np.ones(100)*p_x,label='p(x) = 1')
plt.title('Uniform distribution')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

```python
y = np.exp(x)
plt.hist(y,20,density=True)
#p(y)=(dx/dy)p(x)=(1/y)*(1/10)
y_plot = np.linspace(np.exp(np.min(x)),np.exp(np.max(x)),100)
plt.plot(y_plot,(np.ones(100)*p_x/y_plot),label='p(y) = p(x)/y)')
plt.title('y=exp(x)')
plt.xlabel('y')
plt.ylabel('p(y)')
plt.legend()
plt.show()
```

```python
y = np.log10(x)
plt.hist(y,20,density=True)
#p(y)=(dx/dy)p(x)=(10^y)*ln10
y_plot = np.linspace(math.log10(np.min(x)),math.log10(np.max(x)),len(y))
plt.plot(y_plot,math.log(10)*10**y_plot*p_x,label='p(y) = ln(10)*10^y')
plt.title('y=log10(x)')
plt.xlabel('y')
plt.ylabel('p(y)')
plt.legend()
plt.show()
```

```python
print('log10(mean(x)) = ', math.log10(np.mean(x)))
print('mean(y) = ', np.mean(y))
print('They should not be the same')
```

```python
print('log(median(x) = ', math.log10(statistics.median(x)))
print('median(y) = ', statistics.median(y))
print('They should be the same')
```
