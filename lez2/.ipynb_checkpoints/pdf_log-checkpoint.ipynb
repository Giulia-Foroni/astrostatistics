```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
from numpy import zeros
import math
import statistics
```

Generate $N$ random samples from a uniform distribution [0.1,10]

```python
x_min = 0.1
x_max = 10
N = 5000
x = np.random.uniform(x_min,x_max,N) 
p_x = 1/(x_max-x_min)
```

$p(x)=1/10$ if $x\in[0.1,10]$
\
$p(x)=0$ otherwise

```python
plt.hist(x,20,density=True,label='samples')
plt.plot(np.linspace(np.min(x),np.max(x),100),np.ones(100)*p_x,label='p(x)')
plt.title('Uniform distribution [0.1,10]')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

Let's define a new variable $y$ such that $y=\log_{10}(x)$
\
In general, $p(y)=|\frac{dx}{dy}|*p(x)$. Therefore, the theorical pdf for y is:
\
$p(y)=(10^y)*\ln10*p(x)$

```python
y = np.log10(x)
plt.hist(y,20,density=True,label='samples')
y_plot = np.linspace(math.log10(np.min(x)),math.log10(np.max(x)),len(y))
plt.plot(y_plot,math.log(10)*10**y_plot*p_x,label='p(y)')
plt.title('y=log$_{10}$(x)')
plt.xlabel('y')
plt.ylabel('p(y)')
plt.legend()
plt.show()
```

Let's compute the $\log_{10}$ of the mean of $x$ and the mean of $y$: they should not be the same since the mean depends on the scale of the sample values.

```python
print('log10(mean(x)) = ', math.log10(np.mean(x)))
print('mean(y) = ', np.mean(y))
```

Let's now compute the $\log_{10}$ of the median of $x$ and the median of $y$: they should be the same since the median only depends on the ordering of the samples.

```python
print('log10(median(x)) = ', math.log10(statistics.median(x)))
print('median(y) = ', statistics.median(y))
```
