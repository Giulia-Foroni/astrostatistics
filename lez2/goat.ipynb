```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
from numpy import zeros
```

```python
# goat = 0, car = 1

def random_door (n_door):
    door = np.zeros(n_door,dtype=int)
    car = np.random.randint(0,n_door)
    door[car] = 1
    return door
```

```python
n_door = 3 #numero di porte
d = np.linspace(0,2,n_door,dtype=int) #indici delle n_door porte
N = 10000 #numero di cicli

switch = [] #vittorie o sconfitte del giocatore che cambia porta
confirm = [] #vittorie o sconfitte del giocatore che non cambia porta
new = [] #vittorie o sconfitte del nuovo giocatore
```

```python
for i in range(N):
    door = random_door(n_door)
    g1 = np.random.randint(0,n_door) #scelta del giocatore che cambierà
    host = np.delete(d,g1)
    if door[host[0]]==0:
        g1 = host[1]
        host = host[0]
        g3 = np.delete(d,host)[np.random.randint(0,len(d)-1)] #scelta del nuovo giocatore: una delle porte non aperte dall'host
    else:
        g1 = host[0]
        host = host[1]
        g3 = np.delete(d,host)[np.random.randint(0,len(d)-1)] #scelta del nuovo giocatore: una delle porte non aperte dall'host
    g2 = np.random.randint(0,n_door) #scelta del giocatore che non cambierà
    switch.append(door[g1])
    confirm.append(door[g2])
    new.append(door[g3])

switch = np.array(switch)
confirm = np.array(confirm)
new = np.array(new)
```

```python
print('Probabilità di vittoria:')
p_switch = len(switch[switch==1])/len(switch)
print('Giocatore che cambia scelta: ', p_switch)
p_confirm = len(confirm[confirm==1])/len(confirm)
print('Giocatore che non cambia scelta: ',p_confirm)
p_new = len(new[new==1])/len(new)
print('Nuovo giocatore: ', p_new)
```

```python
def cambio_numero_porte (n_door,n_open): #numero totale di porte, numero di porte aperte
    d = np.arange(0,n_door,1,dtype=int) #indici delle n_door porte
    switch = [] #vittorie o sconfitte del giocatore che cambia porta
    confirm = [] #vittorie o sconfitte del giocatore che non cambia porta
    new = [] #vittorie o sconfitte del nuovo giocatore
    
    for i in range(N):
        door = random_door(n_door) #vettore 0 (capre) e 1 (macchina)
        g1 = np.random.randint(0,n_door) #scelta del giocatore che cambierà
        porte_rimaste = np.delete(d,g1)  #indici porte senza la scelta del giocatore1
        host = [] #porte aperte dall'host
        
        for j in range(n_open):
            h = porte_rimaste[np.random.randint(0,len(porte_rimaste))]
            while door[h]==1 or h in host:
                h = porte_rimaste[np.random.randint(0,len(porte_rimaste))] #scelta dell'host
            host.append(h)

        host = np.array(host)
        g1 = np.delete(d,np.concatenate((np.array([g1]),host)))[np.random.randint(len(d)-1-len(host))] #nuova scleta del giocatore1
        g3 = np.delete(d,host)[np.random.randint(len(d)-len(host))] #scelta del nuovo giocatore: una delle porte non aperte dall'host
        g2 = np.random.randint(0,n_door) #scelta del giocatore che non cambierà
        
        switch.append(door[g1])
        confirm.append(door[g2])
        new.append(door[g3])
    
    switch = np.array(switch)
    confirm = np.array(confirm)
    new = np.array(new)

    p_switch = len(switch[switch==1])/len(switch)       #probabilità di vittoria dei singoli giocatori
    p_confirm = len(confirm[confirm==1])/len(confirm)
    p_new = len(new[new==1])/len(new)

    return [p_switch, p_confirm, p_new]


def p_g1 (n_door,n_open): #probabilità giocatore che cambia scelta
    d = np.arange(0,n_door,1,dtype=int) #indici delle n_door porte
    switch = [] #vittorie o sconfitte del giocatore che cambia porta
        
    for i in range(N):
        door = random_door(n_door) #vettore 0 (capre) e 1 (macchina)
        g1 = np.random.randint(0,n_door) #scelta del giocatore che cambierà
        porte_rimaste = np.delete(d,g1)  #indici porte senza la scelta del giocatore1
        host = [] #porte aperte dall'host
        
        for j in range(n_open):
            h = porte_rimaste[np.random.randint(0,len(porte_rimaste))]
            while door[h]==1 or h in host:
                h = porte_rimaste[np.random.randint(0,len(porte_rimaste))] #scelta dell'host
            host.append(h)

        host = np.array(host)
        g1 = np.delete(d,np.concatenate((np.array([g1]),host)))[np.random.randint(len(d)-1-len(host))] #nuova scleta del giocatore
        switch.append(door[g1])

    switch = np.array(switch)
    p_switch = len(switch[switch==1])/len(switch)  #probabilità di vittoria dei singoli giocatori

    return p_switch

def p_g2 (n_door,n_open): #probabilità giocatore che non cambia scelta
    d = np.arange(0,n_door,1,dtype=int) #indici delle n_door porte
    confirm = [] #vittorie o sconfitte del giocatore che non cambia porta

    for i in range(N):
        door = random_door(n_door) #vettore 0 (capre) e 1 (macchina)
        g2 = np.random.randint(0,n_door) #scelta del giocatore che non cambierà
        confirm.append(door[g2])
    
    confirm = np.array(confirm)
    p_confirm = len(confirm[confirm==1])/len(confirm)

    return p_confirm

def p_g3 (n_door,n_open): #probabilità nuovo giocatore
    d = np.arange(0,n_door,1,dtype=int) #indici delle n_door porte
    new = [] #vittorie o sconfitte del nuovo giocatore
    
    for i in range(N):
        door = random_door(n_door) #vettore 0 (capre) e 1 (macchina)
        g1 = np.random.randint(0,n_door) #scelta del giocatore che cambierà
        porte_rimaste = np.delete(d,g1)  #indici porte senza la scelta del giocatore1
        host = [] #porte aperte dall'host
        
        for j in range(n_open):
            h = porte_rimaste[np.random.randint(0,len(porte_rimaste))]
            while door[h]==1 or h in host:
                h = porte_rimaste[np.random.randint(0,len(porte_rimaste))] #scelta dell'host
            host.append(h)

        host = np.array(host)
        g3 = np.delete(d,host)[np.random.randint(len(d)-len(host))] #scelta del nuovo giocatore: una delle porte non aperte dall'host
       
        new.append(door[g3])
    
    new = np.array(new)

    p_new = len(new[new==1])/len(new)

    return p_new
```
