```python
import sklearn
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import metrics
from astroML.utils import completeness_contamination
from sklearn.metrics import roc_curve
from sklearn.manifold import Isomap

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier

import pandas as pd
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
import seaborn as sns
```

```python
import h5py
f = h5py.File('sample_2e7_design_precessing_higherordermodes_3detectors.h5', 'r')
list(f.keys())
```

```python
d = {i: f[i][:] for i in f.keys()}
data = pd.DataFrame(data=d)
data
```

```python
for i in list(f.keys()):
    print(len(data[i][data[i]=='nan']))
```

First, let's use only the first 5000 rows. Let's split our dataset in test and train sets.

```python
smaller_data = data[:5000]
smaller_data = smaller_data.drop(['det', 'snr'],axis=1)
smaller_labels = data.det[:5000]

x_train, x_test, y_train, y_test = train_test_split(smaller_data, smaller_labels, train_size = 0.8)
```

```python
plt.scatter(smaller_data['z'],smaller_data['mtot'],c=smaller_labels,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])
```

# Logistic Regression

```python
predictions = []
classifier = []

for i in list(f.keys()):
    if i == 'det' or i == 'snr':
        continue
    clf = LogisticRegression(class_weight='balanced',max_iter=1000)
    clf.fit(x_train[i].to_numpy()[:,np.newaxis],y_train)
    y_pred = clf.predict(x_test[i].to_numpy()[:,np.newaxis])
    
    predictions.append(y_pred)
    classifier.append(clf)

completeness, contamination = completeness_contamination(predictions, y_test)
```

```python
x = smaller_data.columns[smaller_data.columns != 'det']
x = x[x != 'snr']

fig, axes = plt.subplots(2,1)
fig.subplots_adjust(hspace=0.6)

axes[0].plot(x,completeness)
axes[0].set_title('completeness')
axes[0].set_ylabel('completeness')
axes[0].tick_params("x", rotation=45)

axes[1].plot(x,contamination)
axes[1].set_title('contamination')
axes[1].set_ylabel('contamination')
axes[1].tick_params("x", rotation=45)

plt.show()
```

```python
for i,j in zip(x,range(len(classifier))):
    scores = classifier[j].predict_proba(x_test[i].to_numpy()[:,np.newaxis])[:,1]
    fpr, tpr, thresh = roc_curve(y_test, scores)
    plt.plot(fpr,tpr,label=i)
    plt.legend()
plt.title('ROC curves')
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.show()
```

The feature 'z' is the one that produces the highest completeness and the lower contamination. Let's train our algorithm using only this feature, then predict the labels in the test set and plot two components to visualize the result.

```python
clf = LogisticRegression(class_weight='balanced',max_iter=1000)
clf.fit(x_train['z'].to_numpy()[:,np.newaxis],y_train)
y_pred = clf.predict(x_test['z'].to_numpy()[:,np.newaxis])
plt.scatter(x_test['z'].to_numpy(),x_test['mtot'].to_numpy(),c=y_pred,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])

xlim = (0, 4)
ylim = (0, 1000)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 51),
                     np.linspace(ylim[0], ylim[1], 51))
Z = clf.predict(xx.ravel()[:, np.newaxis])
Z = Z.reshape(xx.shape)
plt.contour(xx,yy,Z,[0.5],colors='k')

plt.xlabel('z')
plt.ylabel('mtot')
plt.show()
```

# Decision tree

Let's optimize the parameter max_depth using cross-validation.

```python
model = DecisionTreeClassifier()
depth = np.arange(1,40,1)
grid = GridSearchCV(model, {'max_depth': depth}, cv = 5)
grid.fit(x_train['z'].to_numpy()[:,np.newaxis],y_train)
print('Best iperparameter: ', grid.best_params_)
print('Best score: ', grid.best_score_)
```

```python
dtree = DecisionTreeClassifier(max_depth=grid.best_params_['max_depth'])
dtree.fit(x_train['z'].to_numpy()[:,np.newaxis],y_train)
y_pred = dtree.predict(x_test['z'].to_numpy()[:,np.newaxis])
plt.scatter(x_test['z'].to_numpy(),x_test['mtot'].to_numpy(),c=y_pred,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])
plt.xlabel('z')
plt.ylabel('mtot')

xlim = (0, 4)
ylim = (0, 1000)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 51),
                     np.linspace(ylim[0], ylim[1], 51))
Z = dtree.predict(xx.ravel()[:, np.newaxis])
Z = Z.reshape(xx.shape)
plt.contour(xx,yy,Z,[0.5],colors='k')

plt.show()
```

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(max_depth=grid.best_params_['max_depth']), 
    n_estimators=500, max_samples=100, 
    bootstrap=True, n_jobs=-1)
bag_clf.fit(x_train['z'].to_numpy()[:, np.newaxis],y_train)
y_pred = bag_clf.predict(x_test['z'].to_numpy()[:, np.newaxis])
plt.scatter(x_test['z'].to_numpy(),x_test['mtot'].to_numpy(),c=y_pred,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])
plt.xlabel('z')
plt.ylabel('mtot')

xlim = (0, 4)
ylim = (0, 1000)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 51),
                     np.linspace(ylim[0], ylim[1], 51))
Z = bag_clf.predict(xx.ravel()[:, np.newaxis])
Z = Z.reshape(xx.shape)
plt.contour(xx,yy,Z,[0.5],colors='k')

plt.show()
```

# Dimensionality reduction

```python
n_components = 2
embedding = Isomap(n_components=n_components,n_neighbors=5)
X_transformed = embedding.fit_transform(smaller_data)
```

```python
plt.scatter(X_transformed[:,0],X_transformed[:,1],c=smaller_labels,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X_transformed,smaller_labels,test_size=0.2,train_size=0.8)
```

```python
clf = LogisticRegression(class_weight='balanced',max_iter=1000)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred,cmap=plt.colormaps['jet'])
plt.colorbar(ticks=[0,1])

xlim = (-800, 800)
ylim = (-22, 62)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 51),
                     np.linspace(ylim[0], ylim[1], 51))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx,yy,Z,[0.5],colors='k')

plt.show()
```

```python
completeness, contamination = completeness_contamination(y_pred, y_test)
print('Completeness: ', completeness)
print('Contamination', contamination)
```
