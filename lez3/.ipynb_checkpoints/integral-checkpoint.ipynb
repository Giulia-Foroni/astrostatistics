Using Monte Carlo integration, check that $\int_{0}^{+\infty}x^3\exp(-\frac{x^2}{2\sigma^2})\,dx=2\sigma^4$

In our case:
\
$p(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp(-\frac{x^2}{2\sigma^2})$
\
$f(x)=x^3$

```python
N = 100000                                                  # number of samples
sigma = 1
x = np.random.normal(0,sigma,N)                             # samples from p(x)
x = x[x>=0]                                                 # we have to integrate in [0,+inf]
integral = math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3)  # normalization * 1/N * sum(f(x))
print('Numeric integral = ', integral)
print('Analytic integral = ', 2*sigma**4)
```

## How does the error go with the number of samples?

Let's look at how the error scales if we change the number $N$ of samples. In particular we look at the residuals, defined as the absolute value of the difference between the analytic result and the numerical one.

```python
sigma = 1
N = np.linspace(100,10000,1000,dtype=int)  # number of samples
residuals = []
for i in N:
    x = np.random.normal(0,sigma,i)
    x = np.array(x)
    x = x[x>=0] 
    residuals.append(np.absolute(2*sigma**4-math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3)))
plt.plot(N,residuals)
plt.title('Residuals as a function of the number of samples')
plt.xlabel('Number of samples')
plt.ylabel('Residuals')
plt.show()
```

As one would expect, the residuals decrease as we increase the number of samples. So, a bigger $N$ gives a better numerical estimate of the integral.

## For a fixed N, how is the result distributed? How does the distribution change if N increases?

Let's now fix the number of samples $N$ and compute the integral many times to see how the results are distributed. We expect a gaussian centered around the true (analytical) value.

```python
sigma = 1
fig, axes = plt.subplots(2,2)
fig.subplots_adjust(hspace=0.4,wspace=0.4)

integral = []
for i in range(1000):
    x = np.random.normal(0,sigma,50)
    x = np.array(x)
    x = x[x>=0] 
    integral.append(math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3))
integral = np.array(integral)
axes[0,0].hist(integral,50,density=True)
axes[0,0].set_title('N = 50')

integral = []
for i in range(1000):
    x = np.random.normal(0,sigma,100)
    x = np.array(x)
    x = x[x>=0] 
    integral.append(math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3))
integral = np.array(integral)
axes[0,1].hist(integral,50,density=True)
axes[0,1].set_title('N = 100')

integral = []
for i in range(1000):
    x = np.random.normal(0,sigma,200)
    x = np.array(x)
    x = x[x>=0] 
    integral.append(math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3))
integral = np.array(integral)
axes[1,0].hist(integral,50,density=True)
axes[1,0].set_title('N = 200')

integral = []
for i in range(1000):
    x = np.random.normal(0,sigma,500)
    x = np.array(x)
    x = x[x>=0] 
    integral.append(math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3))
integral = np.array(integral)
axes[1,1].hist(integral,50,density=True)
axes[1,1].set_title('N = 500')
plt.show()
```

We see that for increasing values of $N$ the distribution tends to a gaussian. Let's take a big value of $N$, say $N=30000$ and fit the resulting distribution with a gaussian.

```python
N = 30000  
sigma = 1
integral = []
for i in range(1000):
    x = np.random.normal(0,sigma,N)
    x = np.array(x)
    x = x[x>=0] 
    integral.append(math.sqrt(2*math.pi)*sigma/(2*len(x))*sum(x**3))
integral = np.array(integral)
```

```python
def gauss(p,x):
    return 1/(np.sqrt(2*np.pi*p[1]**2))*np.exp(-(x-p[0])**2/2/p[1]**2)
```

```python
a,b,c = plt.hist(integral,50,density=True)        # histogram
mu = np.average(integral)
sigma = np.std(integral)
x_fit = np.linspace(np.min(integral),np.max(integral),100)
plt.plot(x_fit,gauss([mu,sigma],x_fit),label='gaussian')    # gaussian
plt.legend()
plt.title('N = '+str(N))
plt.show()
```

## prova MCMC

```python
def likelihood(p,x):
    return np.prod(gauss(p,x))

def prior(p):
    if p[0]==0:
        return 0
    prior_mu = stats.uniform.pdf(p[0],loc=0,scale=50)
    prior_sigma = stats.uniform.pdf(p[1],loc=0,scale=50)/p[1]
    return prior_mu*prior_sigma
    
def log_posterior(p,x):
    lp = np.log(prior(p))
    ll = np.log(likelihood(p,x))
    if not np.isfinite(lp):
        return -np.inf
    if not np.isfinite(ll):
        return -np.inf
    else:
        return ll+lp
```

```python
ndim = 2
nwalkers = 4
nsteps = 40000
starting_guesses = np.array([-1,0.001])+[.5,1e-2]*np.random.rand(nwalkers, ndim)
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_posterior,args=[integral])
sampler.run_mcmc(starting_guesses, nsteps)
```

```python
fig, axes = plt.subplots(ndim, figsize=(10, 7), sharex=True)
sampler_plot = sampler.get_chain()
labels = [r"$\mu$",r"$\sigma$"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(sampler_plot[:, :, i], alpha=0.3)
    ax.set_xlim(0, len(sampler_plot))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number")
plt.show()
```

```python
a,b,c = plt.hist(integral,50,density=True,label='I') #histogram

medians = []
[medians.append(np.median(sampler_plot[:,:,i])) for i in range(ndim)]
x_fit = np.linspace(np.min(integral),np.max(integral),100)
plt.plot(x_fit,1/(math.sqrt(2*math.pi)*medians[1])*np.exp(-(x_fit-medians[0])**2/(2*medians[1]**2)),label='gaussian') #gaussian

plt.title('Distribution of the integral. N = '+str(N))
plt.legend()
plt.xlabel('I')
plt.show()
```
