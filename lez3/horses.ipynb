```python
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
from scipy.special import factorial
from scipy.stats import poisson
```

First, let's plot our data: number of deaths vs number of groups. We want a normalized distribution, so we divide the number of groups by the total number of groups.

```python
death = np.arange(0,5,1)
number = np.array([109,65,22,3,1])
freq = number/np.sum(number)
plt.plot(death,freq,'.')
plt.title('Data')
plt.show()
```

Now we fit our distribution with a poissonian. We do it for different values of $\mu$.

```python
plt.plot(death,freq,'.')
mu = np.array([0.1,0.6,1,2])
for i in mu:
    dist = poisson(i)
    plt.plot(death,dist.pmf(death),label=r'$\mu$ = '+str(i))
plt.title('Data vs poissonian distribution')
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

It looks like $\mu=0.6$ is the best fit for our data.

```python
plt.plot(death,freq,'.',label='data')
dist = poisson(0.6)
plt.plot(death,dist.pmf(death),label='poissonian')
plt.title('Data vs poissonian distribution')
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.title(r'$\mu$ = 0.6')
plt.legend()
plt.show()
```
