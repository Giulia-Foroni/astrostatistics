```python
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
from scipy.special import factorial
from scipy.stats import poisson
```

```python
death = np.arange(0,5,1)
number = np.concatenate((np.ones(109)*0,np.ones(65)*1,np.ones(22)*2,np.ones(3)*3,np.ones(1)*4))   #data
plt.hist(number,4,density=True,label='Data')
x_plot = np.arange(0,10,1)
mu = 1
dist = poisson(mu)
plt.plot(x_plot,dist.pmf(x_plot),label='Poisson distribution, mu = '+str(mu)) #distribuzione poissoniana per confrontare con l'integrale
plt.title('Number of deaths frequency')
plt.xlabel('Deaths')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
