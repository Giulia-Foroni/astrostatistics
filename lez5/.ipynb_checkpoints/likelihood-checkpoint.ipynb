```python
import numpy as np
import matplotlib as plt
from matplotlib import pyplot as plt
import math
import random
import scipy.stats
from scipy.stats import norm
```

## Homoscedastic errors

```python
mu = 1
sigma = 0.2
N = 100
x = np.random.normal(mu,sigma,N)  #N random numbers that follow a normal distribution with mu = 1 and sigma = 0.2
#x = np.linspace(-1,3,10)
x_plot = np.linspace(-0.5,2.5,500)
likelihood = 1
for i in x:
    distribution = norm(i,sigma)   #normal distribution centered around each random number
    plt.plot(x_plot,distribution.pdf(x_plot))
    likelihood = likelihood*distribution.pdf(x_plot)   #likelihood = product of all the previous distributions
plt.title('Gaussians centered around the random numbers')
plt.xlabel('Position of the quasar')
plt.show()
```

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0],label='likelihood')
plt.xlim(0.8,1.2)
plt.plot(np.ones(10)*x_plot[likelihood == np.max(likelihood)],np.linspace(0,np.max(likelihood),10),'--',label='max of the likelihood')
sigma_mu = 1/np.sqrt(-np.diff(np.log(likelihood[likelihood!=0]),n=2)[0]/(np.diff(x_plot[likelihood!=0])[0]**2))   #estimated value of sigma mu
distribution = norm(x_plot[likelihood == np.max(likelihood)],sigma_mu)
plt.plot(x_plot,distribution.pdf(x_plot)*3.2*10**4,label='Fisher matrix error estimation')  #gaussian centered around the max of the likelihood with sigma = estimated value of sigma mu
#plt.title('Likelihood')
plt.legend()
plt.xlabel('Position of the quasar')
#plt.ylabel('L(mu)')
plt.show()
```

```python
print('mu hat = ', np.sum(x)/N, ' +- ', sigma/math.sqrt(N))
print('Max likelihood = ', x_plot[likelihood == np.max(likelihood)], '+-', sigma_mu)
```

## Heteroscedastic errors

```python
N = 100
mu = 1
sigma = np.random.normal(0.2,0.05,N)  #heteroscedastic errors
x = np.random.normal(mu,sigma,N)   #random numbers
x_plot = np.linspace(-1,3,500)
likelihood = 1
for i,j in zip(x,sigma):
    distribution = norm(i,j)   #normal distribution centered around each random number
    plt.plot(x_plot,distribution.pdf(x_plot))
    likelihood = likelihood*distribution.pdf(x_plot)   #likelihood = product of all the previous distributions
plt.title('Gaussians centered around the random numbers')
plt.xlabel('Position of the quasar')
plt.show()
```

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0],label='likelihood')
plt.xlim(0.8,1.2)
plt.plot(np.ones(10)*x_plot[likelihood == np.max(likelihood)],np.linspace(0,np.max(likelihood),10),'--',label='max of the likelihood')
sigma_mu = 1/np.sqrt(-np.diff(np.log(likelihood[likelihood!=0]),n=2)[0]/(np.diff(x_plot[likelihood!=0])[0]**2))   #estimated value of sigma mu
distribution = norm(x_plot[likelihood == np.max(likelihood)],sigma_mu)
plt.plot(x_plot,distribution.pdf(x_plot)*5.3*10**7,label='Fisher matrix error estimation')  #gaussian centered around the max of the likelihood with sigma obtained differentiating the likelihood
#plt.title('Likelihood')
plt.legend()
plt.xlabel('Position of the quasar')
#plt.ylabel('L(mu)')
plt.show()
```

```python
print('mu hat = ', np.sum(x/sigma**2)/np.sum(1/sigma**2), ' +- ', 1/np.sqrt(np.sum(1/sigma**2)))
print('Max likelihood = ', x_plot[likelihood == np.max(likelihood)], '+-', sigma_mu)
```
