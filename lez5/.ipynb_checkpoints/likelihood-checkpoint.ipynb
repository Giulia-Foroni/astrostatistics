```python
import numpy as np
import matplotlib as plt
from matplotlib import pyplot as plt
import math
import random
import scipy.stats
from scipy.stats import norm
```

## Homoscedastic errors

We measure the position of a quasar $N$ times with the same telescope (i.e. errors are homoscedastic). Let's say the underlying process is Gaussian, the true position is $\mu=1$ (in suitable units) and the erorrs are $\sigma=0.2$.

We treat each of the $N$ measures as the center of a gaussian distribution with $\sigma=0.2$. Each gaussian is the likelihood of getting each of the $N$ measures. Let's plot them.

```python
mu = 1
sigma = 0.2
N = 100
x = np.random.normal(mu,sigma,N)                       # N measures
x_plot = np.linspace(-0.5,2.5,500)
likelihood = 1
for i in x:
    distribution = norm(i,sigma)
    plt.plot(x_plot,distribution.pdf(x_plot))         # likelihood
    likelihood = likelihood*distribution.pdf(x_plot)
plt.xlabel('Position of the quasar')
plt.title('Likelihoods of the N measures')
plt.show()
```

The likelihood of our dataset of $N$ measures is the product of all the likelihoods.

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0])
plt.plot(np.ones(10)*x_plot[likelihood == np.max(likelihood)],np.linspace(0,np.max(likelihood),10),'--',label='max of the likelihood')
plt.xlabel('Position of the quasar')
plt.xlim(0.8,1.2)
plt.title('Likelihood')
plt.legend()
plt.show()
```

Let's compare the maximum of our likelihood with the estimator $\hat\mu=\frac{1}{N}\sum_{i=1}^{N}x_i$.

```python
print('mu hat = ', np.round(np.sum(x)/N,4))
print('Max likelihood = ', np.round(x_plot[likelihood == np.max(likelihood)][0],4))
```

Now, let's look at the Fisher matrix error $F$. We compute it by differentiating our likelihood. Then, we overplot to our likelihood a gaussian with $\mu=max(likelihood)$ and $\sigma=\sqrt{F^{-1}}$ (we have to rescale it by hand).

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0],label='likelihood')
plt.xlim(0.8,1.2)
sigma_mu = 1/np.sqrt(-np.diff(np.log(likelihood[likelihood!=0]),n=2)[0]/(np.diff(x_plot[likelihood!=0])[0]**2))   #estimated value of sigma mu
distribution = norm(x_plot[likelihood == np.max(likelihood)],sigma_mu)
plt.plot(x_plot,distribution.pdf(x_plot)*8.2*10**6,label='Fisher matrix error estimation')  #gaussian centered around the max of the likelihood with sigma = estimated value of sigma mu
plt.legend()
plt.xlabel('Position of the quasar')
plt.show()
```

In our case, $\sigma_\mu=\frac{\sigma}{\sqrt{N}}$. Let's compare this to our estimated $\sigma$.

```python
print('mu hat = ', np.round(np.sum(x)/N,2), ' +- ', sigma/math.sqrt(N))
print('Estimated = ', np.round(x_plot[likelihood == np.max(likelihood)][0],4), '+-', np.round(sigma_mu,4))
```

## Heteroscedastic errors

Now, let's suppose that the errors on our $N$ measures are heteroscedastic and repeat the procedure.

```python
N = 100
mu = 1
sigma = np.random.normal(0.2,0.05,N)  #heteroscedastic errors
x = np.random.normal(mu,sigma,N)   
x_plot = np.linspace(-1,3,500)
likelihood = 1
for i,j in zip(x,sigma):
    distribution = norm(i,j)
    plt.plot(x_plot,distribution.pdf(x_plot))
    likelihood = likelihood*distribution.pdf(x_plot) 
plt.title('Likelihoods of the N measures')
plt.xlabel('Position of the quasar')
plt.show()
```

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0])
plt.plot(np.ones(10)*x_plot[likelihood == np.max(likelihood)],np.linspace(0,np.max(likelihood),10),'--',label='max of the likelihood')
plt.xlabel('Position of the quasar')
plt.xlim(0.8,1.2)
plt.title('Likelihood')
plt.legend()
plt.show()
```

Now $\hat\mu=\frac{\sum_{i=1}^{N}x_i/\sigma_i^2}{\sum_{i=1}^{N}1/\sigma_i^2}$.

```python
print('mu hat = ', np.round(np.sum(x/sigma**2)/np.sum(1/sigma**2),3))
print('Estimated = ', np.round(x_plot[likelihood == np.max(likelihood)][0],3))
```

```python
plt.plot(x_plot[likelihood!=0],likelihood[likelihood!=0],label='likelihood')
plt.xlim(0.8,1.2)
sigma_mu = 1/np.sqrt(-np.diff(np.log(likelihood[likelihood!=0]),n=2)[0]/(np.diff(x_plot[likelihood!=0])[0]**2))   #estimated value of sigma mu
distribution = norm(x_plot[likelihood == np.max(likelihood)],sigma_mu)
plt.plot(x_plot,distribution.pdf(x_plot)*1.3*10**9,label='Fisher matrix error estimation')  #gaussian centered around the max of the likelihood with sigma = estimated value of sigma mu
plt.legend()
plt.xlabel('Position of the quasar')
plt.show()
```

```python
print('mu hat = ', np.round(np.sum(x/sigma**2)/np.sum(1/sigma**2),3), ' +- ', np.round(1/np.sqrt(np.sum(1/sigma**2)),3))
print('Estimated= ', np.round(x_plot[likelihood == np.max(likelihood)][0],3), '+-', np.round(sigma_mu,3))
```
