```python
import numpy as np
import scipy.stats
import pylab as plt
from matplotlib import pyplot as plt
from scipy import stats
import random
import math
import sklearn
from sklearn.mixture import GaussianMixture
```

```python
#plot of the data
data = np.load("formationchannels.npy")
a,b,c=plt.hist(data,100,density=True)
plt.title('Dataset')
plt.show()
```

```python
#compute the value of AIC for different values of N (number of gaussians)
aic = []
N = np.linspace(1,10,10,dtype=int)
for i in N:
    gm = GaussianMixture(n_components=i).fit(data)
    aic.append(gm.aic(data))
plt.plot(N,aic,'.')
plt.title('N vs AIC')
plt.xlabel('N (number of gaussians)')
plt.ylabel('AIC')
plt.show()
print('Minimum value for AIC: N = ', N[np.array(aic)==np.min(np.array(aic))][0])
```
